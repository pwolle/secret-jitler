The [Baconian theory](https://en.wikipedia.org/wiki/Baconian_theory_of_Shakespeare_authorship) holds that Sir Francis Bacon is the author of Shakespeare’s plays. We want to perform a very simple stylistic analysis between Shakespeare’s play Macbeth and Bacon’s New Atlantis.
```python
import urllib.request

atlantis_text = urllib.request \
    .urlopen("https://ia801309.us.archive.org/24/items/newatlantis02434gut/nwatl10.txt") \
    .read() \
    .decode("unicode_escape")

atlantis_text = get_speaker_text(atlantis_text)
atlantis_text = normalize_text(atlantis_text)
atlantis_text = tokenize_text(atlantis_text)
atlantis_text = remove_stop_words(atlantis_text)


macbeth_text = urllib.request \
    .urlopen("https://ia802707.us.archive.org/1/items/macbeth02264gut/0ws3410.txt") \
    .read() \
    .decode("utf-8")

macbeth_text = get_speaker_text(macbeth_text)
macbeth_text = normalize_text(macbeth_text)
macbeth_text = tokenize_text(macbeth_text)
macbeth_text = remove_stop_words(macbeth_text)
```

```python
common_words = set(atlantis_text) & set(macbeth_text)

macbeth_freq = {word: 0 for word in common_words}

for word in macbeth_text:
    if word in macbeth_freq:
        macbeth_freq[word] += 1


atlantis_freq = {word: 0 for word in common_words}

for word in atlantis_text:
    if word in atlantis_freq:
        atlantis_freq[word] += 1


coocurrencies = []

for word in common_words:
    doc1_feq = macbeth_freq[word]
    doc2_feq = atlantis_freq[word]
    coocurrencies.append((word, doc1_feq, doc2_feq, doc1_feq + doc2_feq))


coocurrencies.sort(key=lambda x: x[3], reverse=True)
```

```python
import csv

with open("coocurrencies.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["word", "macbeth_freq", "atlantis_freq", "coocurrencies"])
    writer.writerows(coocurrencies)
```
